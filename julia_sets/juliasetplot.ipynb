```python
# load data from results for strong scaling : runtime with 1 thread/runtime with N threads
speedup = [1, 2, 4, 8, 16, 32]
speedup = np.array(speedup)
ranks = [ 1, 2, 4, 8, 16, 32 ]
ranks = np.array(ranks)
decomp = [145.04, 71.87, 50.3435, 27.573, 16.7267, 9.356]
decomp = np.array(decomp)
mslave = [ 146.42, 148.59, 54.435, 23.21, 15.437, 15.42]
mslave = np.array(mslave)
```

```python
speed = 145.5
decomp1 = speed/decomp
decomp1 
```

```python
mslave1 = speed / mslave
mslave1
```

```python
fig, ax = plt.subplots()
ax.plot(ranks, mslave1, label="master-Slave")
ax.plot(ranks, decomp1, label="S.Decomposition")
ax.plot(ranks,speedup, label="perfect" )
legend = ax.legend(loc="upper left", shadow=True, fontsize='x-small')
plt.xlabel("Number of MPI Rank Elements")
plt.ylabel("speedup")

#plt.pl.xscale(linear, perf, block, noblock)
plt.show()
```
